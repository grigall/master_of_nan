```python
import os
import re
import numpy as np
import pandas as pd
import random

data = pd.read_csv('Employees.csv', sep='\t')
data.head()
```

```python
course_names = ['Chemtrails and Tin-foil Hats: How to Turn Conspiracy Theories into Selling Points',
'Is 5G a myth? (And Other Questions You May Hear)',
'How to Sell Phone Warranties',
 'Smart Phones and Dumb People: Know Your Audience',
 'What Does this Button Do?',
 'Emotional Intelligence for Managers',
 'Emotional Intelligence for Employees'
]

df_cn = pd.DataFrame(course_names, columns=['Course_Name'])
df_cn['Course_ID'] = np.arange(100, (len(df_cn)*100+100),100)
df_cn['Course_Length'] = [random.randint(1,4) for i in range(0,len(df_cn))]
df_cn
```

```python
course_history = []
for i in range(0,len(data)):
    course_list = random.choices(df_cn.Course_ID, k=random.randint(1,10))
    eid = data.EID[i]
    course_history.append([eid, course_list])

df_ch = pd.DataFrame(course_history, columns=['EID', 'Course_List'])
df_ch.head()
```

```python
def date_str_gen(year_start, year_end):
    year = random.randint(year_start, year_end)

    const_day_range = [31,28,31,30,31,30,31,31,30,31,30,31]
    
    str_yr = str(year)
    rand_mo = random.randint(1,12)
    
    rand_day = random.randint(1,const_day_range[(rand_mo-1)])
    
    if len(str(rand_mo)) == 1:
        rand_mo = '0' + str(rand_mo)
    else:
        rand_mo = str(rand_mo)
        
    if len(str(rand_day)) == 1:
        rand_day = '0' + str(rand_day)
    else:
        rand_day = str(rand_day)
        
    iso_date = str_yr + '-' + rand_mo + '-' + rand_day
    
    return iso_date
```

```python
new_date = date_str_gen(2017,2022)
print(new_date)
```

```python
all_dates = []
for person in range(0,len(df_ch)):
    course_dates = []
    for course in df_ch.Course_List[person]:
        course_dates.append(date_str_gen(2010,2022))
    all_dates.append([df_ch.EID[person], course_dates])
    
df_dates = pd.DataFrame(all_dates, columns=['EID', 'Course_Dates'])
df_dates.head()
```

```python
df_blank = pd.DataFrame(None)
```

```python
for cn in range(0,len(df_ch)):
    df_new = pd.DataFrame([df_ch.Course_List[cn], df_dates.Course_Dates[cn]])
    df_t = df_new.transpose()
    df_t.columns = ['Course_ID', 'Course_Date']
    df_t['EID'] = df_ch.EID[cn]
    df_blank = pd.concat([df_blank, df_t], axis=0)
```
